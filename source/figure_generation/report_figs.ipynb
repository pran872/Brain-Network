```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def line_graph(x, ys, xlabel, ylabel, legend, save_pt=None, ylims=None, loc=None, yaxis=True, title=None):
    plt.rcParams.update({
        "text.usetex": False,
        "font.family": "serif",
        "font.serif": ["cmr10", "Computer Modern", "DejaVu Serif"],  # cmr10 closely matches LaTeX
        "axes.labelsize": 14,
        "font.size": 14,
        "legend.fontsize": 12,
        "xtick.labelsize": 12,
        "ytick.labelsize": 12,
        "axes.linewidth": 1,
        "lines.markersize": 6,
        "lines.linewidth": 2,
        "legend.frameon": False,
        "legend.handlelength": 2,
    })
    if ylims is not None:
        plt.ylim(ylims)
    for y in ys:
        plt.plot(x, y, "-s")
    if title:
        plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(legend, loc=loc)
    plt.gca().axes.get_yaxis().set_visible(yaxis)
    plt.tight_layout()
    if save_pt is not None:
        plt.savefig(f"final_figures/{save_pt}", dpi=300)
    plt.show()
```

```python
models = ['Zoom', 'Zoom+Focus', 'ResNet-18', 'Hybrid Baseline']
attacks = ['FGSM', 'PGD']  # Gaussian removed

data = {
    'CIFAR-10 Scratch':      [[50.88, 36.71], [53.02, 41.72], [47.70, 38.81], [50.09, 36.07]],
    'CIFAR-10 Pretrained':   [[48.21, 40.89], [26.55, 41.31], [53.19, 43.64], [17.76, 31.10]],
    'Dogs Scratch':          [[1.76, 0.91], [2.04, 1.54], [1.21, 3.19], [1.60, 0.72]],
    'Dogs Pretrained':       [[2.68, 0.43], [2.79, 0.76], [3.62, 1.98], [2.73, 0.79]]
}

plt.rcParams.update({'font.size': 14})

fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharey='row')
fig.subplots_adjust(hspace=0.3, wspace=0.2)

titles = ['CIFAR-10 (Scratch)', 'CIFAR-10 (Pretrained)', 'Stanford Dogs (Scratch)', 'Stanford Dogs (Pretrained)']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

x = np.arange(len(attacks)) 
# x = np.linspace(1, len(attacks) - 0.5, len(attacks))
width = 0.2 # Adjust spacing since fewer bars per group

for i, (key, ax) in enumerate(zip(data.keys(), axs.flatten())):
    scores = data[key]
    transposed = list(zip(*scores))  # Now [attack][model]

    ax.set_axisbelow(True)
    ax.grid(True, axis='y', linestyle='--', alpha=0.7)

    for j in range(len(models)):
        ax.bar(x + j * width, [transposed[k][j] for k in range(len(attacks))],
               width, label=models[j], color=colors[j])

    ax.set_title(titles[i], fontsize=17)
    group_center = x + (len(models) - 1) * width / 2
    ax.set_xticks(group_center)
    ax.set_xticklabels(attacks, rotation=0, fontsize=15)
    ax.tick_params(axis='y', labelsize=13)

    if i in [1, 3]:
        ax.tick_params(labelleft=False)
    else:
        ax.set_ylabel('Accuracy (%)', fontsize=14)

    ax.set_ylim(0, 60 if i in [0, 1] else 10)

    label = ['A', 'B', 'C', 'D'][i]
    ax.text(-0.1 if i in [0, 2] else -0.01, 1.2, label, transform=ax.transAxes,
            fontsize=18, va='top', ha='left', fontweight='black', fontfamily='serif')

# Global legend
handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center',
           ncol=len(models), fontsize=19, bbox_to_anchor=(0.5, -0.06))

plt.tight_layout()
# plt.savefig("final_figures/robustness.pdf", dpi=300, bbox_inches='tight', pad_inches=0.2)
plt.show()
```

```python
from matplotlib.patches import Patch

# Labels
methods = ['Zoom', 'Zoom+Focus', 'ResNet-18', 'Hybrid Baseline']
severities = ['s1', 's2', 's3', 's4', 's5']
x = np.arange(len(severities)) * 1.5
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Data
corruptions = {
    'Digital': [
        [76.18, 73.83, 72.43, 66.71, 56.98],
        [75.78, 74.04, 73.70, 69.99, 60.82],
        [75.01, 72.34, 70.36, 63.63, 51.92],
        [76.07, 73.83, 72.39, 66.94, 57.46]
    ],
    'Noise': [
        [83.74, 69.45, 54.16, 41.33, 30.15],
        [87.11, 77.48, 65.36, 53.26, 42.39],
        [84.49, 70.56, 54.68, 40.53, 29.29],
        [83.14, 69.64, 55.30, 43.70, 33.71]
    ],
    'Blur': [
        [83.61, 80.67, 76.36, 66.75, 53.96],
        [87.65, 86.08, 83.89, 78.31, 72.64],
        [83.34, 79.63, 74.11, 65.83, 54.09],
        [83.13, 80.17, 75.50, 65.96, 52.63]
    ],
    'Weather': [
        [77.32, 74.22, 72.47, 71.94, 65.94],
        [77.11, 75.07, 73.58, 73.07, 69.23],
        [75.90, 72.58, 70.05, 68.98, 62.05],
        [77.02, 73.83, 72.15, 71.65, 65.69]
    ]
}

plt.rcParams.update({'font.size': 14})
# Create 2x2 grid of plots
fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharey='row', constrained_layout=True)

titles = list(corruptions.keys())
for i, (key, ax) in enumerate(zip(titles, axs.flatten())):
    data = corruptions[key]
    for j in range(len(methods)):
        ax.bar(x + j * 0.2, data[j], width=0.2, color=colors[j], label=methods[j] if i == 0 else "")
    
    if i % 2 == 0:
        ax.set_ylabel("Avg. accuracy (%)", fontsize=14)
    else:
        ax.tick_params(labelleft=False)

    ax.set_title(key, fontsize=17)
    ax.set_xticks(x + 0.3)
    ax.set_xticklabels(range(1, 6), fontsize=13)
    ax.set_xlabel("Severity Level", fontsize=15)
    ax.set_ylim(0, 100)

    # label = ['A', 'B', 'C', 'D'][i]
    # ax.text(-0.1 if i in [0, 2] else -0.01, 1.2, label, transform=ax.transAxes,
    #         fontsize=18, va='top', ha='left', fontweight='black', fontfamily='serif')

# Custom legend (centered below)
handles = [Patch(color=colors[i], label=methods[i]) for i in range(len(methods))]
fig.legend(handles=handles, loc='lower center', ncol=4, fontsize=15, bbox_to_anchor=(0.5, -0.06))

# Save or show
plt.savefig("final_figures/hendrycks_tests.pdf", bbox_inches='tight')
plt.show()
```
